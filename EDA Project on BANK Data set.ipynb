```python
# convert minutes into seconds
duration_sec = []
for i in bank1.duration:
   if 'min' in i:
       m = i.split(' ')
       m = float(m[0])*60
       duration_sec.append(m)
   else:
       a = i.split(' ')
       duration_sec.append(a[0])
     
     
bank1['duration']=duration_sec
```

```python
bank1[['job','edu']] = bank1['jobedu'].str.split(',',1,expand=True)
bank1 = bank1.drop('jobedu',axis =1)
```

```python
bank1 = bank1.drop('poutcome',axis =1)
```

```python
bank1.dtypes
```

```python
bank1['month'].mode()
```

```python
bank1['month'].fillna(('may, 2017'),inplace=True)
```

```python
bank1['month'].value_counts()
```

```python
bank1['age'].mode()
```

```python
bank1['age'].fillna(('32'),inplace=True)
```

```python
bank1['age'].value_counts()
```

```python
bank1['response'].mode()
```

```python
bank1['response'].fillna(('0.0'),inplace=True)
bank1['response'].value_counts()
```

```python
bank1.isnull().sum()
```

```python
bank1.dtypes
```

```python
bank1['salary'].replace('?',np.NaN,inplace=True)
bank1['salary'].value_counts()
```

```python
bank1[["campaign","previous","day","balance","age","salary","duration"]]=bank1[["campaign","previous","day","balance","age","salary","duration"]].apply(pd.to_numeric)
```

```python
bank1.dtypes
```

```python
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(10,5))
axes[0].boxplot(bank1['age'])
axes[0].set_title('age')
axes[1].boxplot(bank1['balance'])
axes[1].set_title('balance')
axes[2].boxplot(bank1['previous'])
axes[2].set_title('previous')
axes[3].boxplot(bank1['campaign'])
axes[3].set_title('campaign')
axes[3].boxplot(bank1['salary'])
axes[3].set_title('salary')
axes[3].boxplot(bank1['duration'])
axes[3].set_title('duration')

fig.savefig("boxplots.png")
```

```python
fig, ax = plt.subplots(figsize=(10,8))
bank1.groupby(['duration']).count()[['age','salary']].plot(ax=ax)
```
